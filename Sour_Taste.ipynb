```python
# Initial States
cs_ion = 1
asic1_prot = 1
asic2_prot = 1
asic3_prot = 0

# Create trace arrays for ASIC1, ASIC2, ASIC3
trace = [[], [], []]  # Each sublist will store values for ASIC1, ASIC2, ASIC3

# while loop 
time = 0
while time <= 100:
    # Trace values to the array
    nodes = (asic1_prot, asic2_prot, asic3_prot)  # Only tracing proteins
    for x in range(0, len(nodes)):
        trace[x].append(nodes[x])

    # Update Rules
    asic1_rule = asic2_prot and (not asic3_prot) and (not cs_ion)
    asic3_rule = asic2_prot and (not asic1_prot)

    # Update values
    asic1_prot = int(asic1_rule)
    asic3_prot = int(asic3_rule)

    # Move forward in time
    time += 1

# Final Output
print("Trace for ASIC1:", trace[0])
print("Trace for ASIC2:", trace[1])
print("Trace for ASIC3:", trace[2])
```

```python
'''
Building all together
'''


import matplotlib.pyplot as plt

# Initial States
cs_ion = 1
asic1 = 1
asic2 = 1
asic3 = 0

# Storage for plotting
time = [0]
asic1_trace = [asic1]

# Track current state
current_state = (asic1, asic3)
steps = 0

while True:
    # Update rules
    new_asic1 = int(asic2 and (not asic3) and (not cs_ion))
    new_asic3 = int(asic2 and (not asic1))

    # Check for steady state
    new_state = (new_asic1, new_asic3)
    if new_state == current_state:
        break

    # Update states
    asic1, asic3 = new_state
    current_state = new_state
    steps += 1

    # Record values
    time.append(steps)
    asic1_trace.append(asic1)

# Plotting
fig, ax = plt.subplots()
ax.plot(time, asic1_trace, marker='o')
ax.set(xlabel='Time Step', ylabel='ASIC1 Value', title='ASIC1 Activity Over Time')
ax.grid()
plt.show()
```
