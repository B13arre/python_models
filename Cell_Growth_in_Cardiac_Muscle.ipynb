```python
# Initial States
igf1 = 1
edn1 = 1
pin1 = 0
mtor = 0

# Time step initialization
time = 0

# Run simulation for 100 steps
while time <= 100:
    # Update Rules
    edn1_rule = igf1
    pin1_rule = igf1
    mtor_rule = igf1 or edn1

    # Update values
    edn1 = int(edn1_rule)
    pin1 = int(pin1_rule)
    mtor = int(mtor_rule)

    # Move forward in time
    time += 1

# Final state output
print("Final State:")
print("IGF1 =", igf1)
print("EDN1 =", edn1)
print("PIN1 =", pin1)
print("mTOR =", mtor)
```

```python
import matplotlib.pyplot as plt

# Initial States
igf1 = 1      # ON
edn1 = 1      # ON
pin1 = 0      # OFF
mtor = 0      # OFF

# Trace arrays
timesteps = 10
time = []
igf1_trace = []
edn1_trace = []
pin1_trace = []
mtor_trace = []

# Run simulation
for t in range(timesteps):
    # Record current state
    time.append(t)
    igf1_trace.append(igf1)
    edn1_trace.append(edn1)
    pin1_trace.append(pin1)
    mtor_trace.append(mtor)

    # Update rules
    edn1 = igf1
    pin1 = igf1
    mtor = igf1 or edn1

# Plot results
plt.figure(figsize=(10, 5))
plt.plot(time, igf1_trace, label="IGF1")
plt.plot(time, edn1_trace, label="EDN1")
plt.plot(time, pin1_trace, label="PIN1")
plt.plot(time, mtor_trace, label="mTOR")
plt.xlabel("Time Step")
plt.ylabel("Boolean State (0 or 1)")
plt.title("Boolean Simulation of Cardiac Cell Growth Pathway")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
